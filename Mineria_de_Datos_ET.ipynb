## **Fase IV. Data Modeling**



```python
df=seleccionados.copy()
```

```python
X = df.drop(['RainTomorrow'], axis=1)
Y = df['RainTomorrow']
```



> Separamos la data con el fin de lograr hacer un análisis en relación a RainTomorrow para poder solventar la pregunta de si llovera el dia de mañana o no en las principales ciudades de la isla de Tasmania



```python
random_state = 42
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.33,random_state=42,shuffle= True)
print('Los datos para entrenamiento son :',X_train.shape)
print('Los datos para prueba son :',X_test.shape)
print('Los etiquetados para entrenamiento son :',y_train.shape)
print('Los etiquetados para prueba son :',y_test.shape)
```

```python
#Instanciar el modelo Árbol
arbol = DecisionTreeClassifier(criterion='entropy',
                                             min_samples_split=20,
                                             min_samples_leaf=5,
                                             max_depth = 7,
                                             class_weight="balanced",
                                             random_state=random_state)
#Entrenamiento del modelo
arbol = arbol.fit(X_train,y_train)
#Predicción del modelo
y_pred_arbol = arbol.predict(X_test)
```

```python
#Instanciar el modelo SVC
svc = SVC(gamma=0.001, C=100, kernel='linear', random_state=123)
#Entrenamiento del modelo
svc = svc.fit(X_train,y_train)
#Predicción del modelo
y_pred_svc = svc.predict(X_test)
```

```python
#Instanciar el modelo KNC
knc = KNeighborsClassifier(algorithm='kd_tree',n_neighbors=6,weights='uniform',leaf_size=6)
#Entrenamiento del modelo
knc = knc.fit(X_train,y_train)
#Predicción del modelo
y_pred_knc = knc.predict(X_test)
```

```python
#Instanciar el modelo RFC
rfc = RandomForestClassifier(criterion='entropy',
                                             min_samples_split=20,
                                             min_samples_leaf=5,
                                             max_depth = 7,
                                             class_weight="balanced",
                                             random_state=random_state)
#Entrenamiento del modelo
rfc = rfc.fit(X_train,y_train)
#Predicción del modelo
y_pred_rfc = rfc.predict(X_test)
```

```python
#Instanciar el modelo RL
rl = LogisticRegression( solver='newton-cg')
#Entrenamiento del modelo
rl = rl.fit(X_train,y_train)
#Predicción del modelo
y_pred_rl = rl.predict(X_test)
```

```python
print("----------------------------------------------------------------------------")
print("----------------------Árbol de clasificación--------------------------------")
print(confusion_matrix(y_test,y_pred_arbol))
confusion_arbol=confusion_matrix(y_test,y_pred_arbol)
vn_arbol,fp_arbol,fn_arbol,vp_arbol = confusion_arbol.ravel()
print("----------------------------------------------------------------------------")
print("----------------------SVC---------------------------------------------------")
print(confusion_matrix(y_test,y_pred_svc))
confusion_svc=confusion_matrix(y_test,y_pred_svc)
vn_svc,fp_svc,fn_svc,vp_svc = confusion_svc.ravel()
print("----------------------------------------------------------------------------")
print("----------------------KNC---------------------------------------------------")
print(confusion_matrix(y_test,y_pred_knc))
confusion_knc=confusion_matrix(y_test,y_pred_knc)
vn_knc,fp_knc,fn_knc,vp_knc = confusion_knc.ravel()
print("----------------------------------------------------------------------------")
print("----------------------RFC---------------------------------------------------")
print(confusion_matrix(y_test,y_pred_rfc))
confusion_rfc=confusion_matrix(y_test,y_pred_rfc)
vn_rfc,fp_rfc,fn_rfc,vp_rfc = confusion_rfc.ravel()
print("----------------------------------------------------------------------------")
print("----------------------Regresión L-------------------------------------------")
print(confusion_matrix(y_test,y_pred_rl))
confusion_rl=confusion_matrix(y_test,y_pred_rl)
vn_rl,fp_rl,fn_rl,vp_rl = confusion_rl.ravel()
print("----------------------------------------------------------------------------")
```

###Métricas de rendimiento

```python
print("----------------------------------------------------------------------------")
print("----------------------Árbol de clasificación--------------------------------")
print('Exactitud del árbol:', (vp_arbol+vn_arbol)/(vp_arbol+fp_arbol+fn_arbol+vn_arbol))
print('Precisión del árbol:', (vp_arbol)/(vp_arbol+fp_arbol))
print('Sensibilidad del árbol:', (vp_arbol)/(vp_arbol+fn_arbol))
print('Especifidad del árbol:', (vn_arbol)/(vn_arbol+fp_arbol))
print('Puntaje F1 del árbol:', f1_score(y_test, y_pred_arbol))
print("----------------------------------------------------------------------------")
print("----------------------SVC---------------------------------------------------")
print('Exactitud de SVC:', (vp_svc+vn_svc)/(vp_svc+fp_svc+fn_svc+vn_svc))
print('Precisión de SVC:', (vp_svc)/(vp_svc+fp_svc))
print('Sensibilidad de SVC:', (vp_svc)/(vp_svc+fn_svc))
print('Especifidad de SVC:', (vn_svc)/(vn_svc+fp_svc))
print('Puntaje F1 de SVC:', f1_score(y_test, y_pred_svc))
print("----------------------------------------------------------------------------")
print("----------------------KNC---------------------------------------------------")
print('Exactitud de KNC:', (vp_knc+vn_knc)/(vp_knc+fp_knc+fn_knc+vn_knc))
print('Precisión de KNC:', (vp_knc)/(vp_knc+fp_knc))
print('Sensibilidad de KNC:', (vp_knc)/(vp_knc+fn_knc))
print('Especifidad de KNC:', (vn_knc)/(vn_knc+fp_knc))
print('Puntaje F1 de KNC:', f1_score(y_test, y_pred_knc))
print("----------------------------------------------------------------------------")
print("----------------------RFC---------------------------------------------------")
print('Exactitud de RFC:', (vp_rfc+vn_rfc)/(vp_rfc+fp_rfc+fn_rfc+vn_rfc))
print('Precisión de RFC:', (vp_rfc)/(vp_rfc+fp_rfc))
print('Sensibilidad de RFC:', (vp_rfc)/(vp_rfc+fn_rfc))
print('Especifidad de RFC:', (vn_rfc)/(vn_rfc+fp_rfc))
print('Puntaje F1 de RFC:', f1_score(y_test, y_pred_rfc))
print("----------------------------------------------------------------------------")
print("----------------------Regresión L-------------------------------------------")
print('Exactitud de RFC:', (vp_rl+vn_rl)/(vp_rl+fp_rl+fn_rl+vn_rl))
print('Precisión de RFC:', (vp_rl)/(vp_rl+fp_rl))
print('Sensibilidad de RFC:', (vp_rl)/(vp_rl+fn_rl))
print('Especifidad de RFC:', (vn_rl)/(vn_rl+fp_rl))
print('Puntaje F1 de RFC:', f1_score(y_test, y_pred_rl))
print("----------------------------------------------------------------------------")
```

###ROC - AUC

```python
contrario = [(0 if y_test.values[pivote] == 1 else 1) for pivote in range(len(y_test))]
base = [0 for pivote in range(len(y_test))]
puntaje_base = roc_auc_score(y_test,base)
puntaje_contrario = roc_auc_score(y_test,contrario)
# Calcular el porcentaje ROC-AUC-arbol
print("----------------------------------------------------------------------------")
print("----------------------Árbol-------------------------------------------------")
puntaje_roc_arbol = roc_auc_score(y_test,y_pred_arbol)
print('AUC score: ',roc_auc_score(y_test,y_pred_arbol))
print("----------------------------------------------------------------------------")
print("----------------------SVC---------------------------------------------------")
puntaje_roc_svc = roc_auc_score(y_test,y_pred_svc)
print('AUC score: ',roc_auc_score(y_test,y_pred_svc))
print("----------------------------------------------------------------------------")
print("----------------------KNC---------------------------------------------------")
puntaje_roc_knc = roc_auc_score(y_test,y_pred_knc)
print('AUC score: ',roc_auc_score(y_test,y_pred_knc))
print("----------------------------------------------------------------------------")
print("----------------------RFC---------------------------------------------------")
puntaje_roc_rfc = roc_auc_score(y_test,y_pred_rfc)
print('AUC score: ',roc_auc_score(y_test,y_pred_rfc))
print("----------------------------------------------------------------------------")
print("----------------------Regresión L-------------------------------------------")
puntaje_roc_rl = roc_auc_score(y_test,y_pred_rl)
print('AUC score: ',roc_auc_score(y_test,y_pred_rl))
print("----------------------------------------------------------------------------")
```
